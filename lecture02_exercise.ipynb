## Setup — Run this cell first


```python
import pandas as pd
import numpy as np

# World Happiness Report 2023 — representative data
# Source: https://www.kaggle.com/datasets/ajaypalsinghlo/world-happiness-report-2023

df = pd.read_csv('../data/world_happiness_2023.csv')
df.columns = ['Country','Region','Happiness_Score','GDP','Social_Support',
              'Life_Expectancy','Freedom','Generosity','Corruption']


print(f"Dataset: {len(df)} countries, {len(df.columns)} columns")
print(df.head())
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Explore the dataset before you start
print("Regions in dataset:")
print(df['Region'].value_counts())
print("\nScore range:", df['Happiness_Score'].min(), "–", df['Happiness_Score'].max())
print("\nBottom 10 countries:")
print(df.nsmallest(10, 'Happiness_Score')[['Country','Region','Happiness_Score']])
```

## Task 1 — Regional Comparison Bar Chart

**What to build:** A horizontal bar chart showing the **average happiness score by region**, sorted from highest to lowest.

**Requirements:**
- Horizontal orientation (category names are long)
- Sorted by score, descending (so the happiest region is at the top)
- Zero baseline on x-axis
- At least one design choice that goes beyond the Plotly default (colour, annotation, labels, etc.)
- An insight title that answers: *which region stands out and why does it matter?*

**Hint:** Use `df.groupby('Region')['Happiness_Score'].mean()` to compute the averages.


```python
# Task 1: Regional comparison bar chart
# -------------------------------------

# Step 1: Compute average happiness score by region
region_avg = (df.groupby('Region')['Happiness_Score']
              .mean()
              .reset_index()
              .sort_values('Happiness_Score'))  # sort for horizontal bar

print(region_avg)

# Step 2: Build your chart
# YOUR CODE HERE

import pandas as pd
import plotly.express as px

# Average happiness score by region
region_avg = (
    df.groupby('Region')['Happiness_Score']
      .mean()
      .sort_values(ascending=False)
      .reset_index()
)

# Highlight the happiest region
colors = ['#1f77b4' if i != 0 else '#ff7f0e'
          for i in range(len(region_avg))]

fig = px.bar(
    region_avg,
    x='Happiness_Score',
    y='Region',
    orientation='h',
    text='Happiness_Score'
)

fig.update_traces(
    marker_color=colors,
    texttemplate='%{text:.2f}',
    textposition='outside'
)

fig.update_layout(
    title={
        'text': 'Western Europe leads global happiness, averaging over 6.9 points across countries',
        'x': 0.5
    },
    xaxis_title='Average Happiness Score',
    yaxis_title='',
    template='simple_white',
    showlegend=False,
    height=600
)

# Zero baseline (required)
fig.update_xaxes(range=[0, region_avg['Happiness_Score'].max() + 1])

# Put highest region at the top
fig.update_yaxes(categoryorder='array',
                 categoryarray=region_avg['Region'])

fig.show()
```

## Task 2 — Bottom vs. Top: A Contrast Story

**What to build:** A bar chart that highlights the **gap between the happiest and least happy countries**, focusing on a specific insight.

**Requirements:**
- Show the **top 8 AND bottom 8 countries** together (16 bars total)
- Use **colour** to distinguish the two groups (not Plotly's default rainbow)
- Add a **visual separator or annotation** that emphasises the gap
- Insight title that tells the story of the gap

**Hint:** Use `pd.concat([df.nlargest(8,'Happiness_Score'), df.nsmallest(8,'Happiness_Score')])` to get both groups.

**Stretch goal:** Add a vertical reference line showing the global average.


```python
# Task 2: Top 8 vs. Bottom 8 contrast
# ------------------------------------

# Step 1: Get top and bottom countries
top8 = df.nlargest(8, 'Happiness_Score').copy()
top8['Group'] = 'Top 8'
bottom8 = df.nsmallest(8, 'Happiness_Score').copy()
bottom8['Group'] = 'Bottom 8'

combined = pd.concat([bottom8, top8]).sort_values('Happiness_Score')
global_avg = df['Happiness_Score'].mean()
print(f"Global average: {global_avg:.2f}")

# Step 2: Build your chart
# YOUR CODE HERE

import pandas as pd
import plotly.express as px

# Top 8 and Bottom 8 countries
top8 = df.nlargest(8, 'Happiness_Score').copy()
top8['Group'] = 'Top 8'

bottom8 = df.nsmallest(8, 'Happiness_Score').copy()
bottom8['Group'] = 'Bottom 8'

comparison = pd.concat([top8, bottom8])

# Sort for display
comparison = comparison.sort_values('Happiness_Score', ascending=True)

# Global average
global_avg = df['Happiness_Score'].mean()

# Custom colors (not Plotly defaults)
color_map = {
    'Top 8': '#2E8B57',      # green
    'Bottom 8': '#C0392B'    # red
}

fig = px.bar(
    comparison,
    x='Happiness_Score',
    y='Country',
    color='Group',
    orientation='h',
    text='Happiness_Score',
    color_discrete_map=color_map
)

fig.update_traces(
    texttemplate='%{text:.2f}',
    textposition='outside'
)

fig.update_layout(
    title={
        'text': 'The happiest countries score more than twice as high as the least happy nations',
        'x': 0.5
    },
    xaxis_title='Happiness Score',
    yaxis_title='',
    template='simple_white',
    legend_title='',
    height=700
)

# Zero baseline (required)
fig.update_xaxes(range=[0, comparison['Happiness_Score'].max() + 1])

# Stretch goal: global average reference line
fig.add_vline(
    x=global_avg,
    line_dash='dash',
    line_color='black'
)

fig.add_annotation(
    x=global_avg,
    y=1.02,
    yref='paper',
    text=f'Global Average: {global_avg:.2f}',
    showarrow=False
)

# Visual annotation emphasizing the gap
fig.add_annotation(
    x=5.2,
    y=7.5,
    text='Large happiness gap between the top and bottom countries',
    showarrow=True,
    arrowhead=2
)

fig.show()
```

If you finish both tasks with time to spare, try this:

**Task 3 (stretch):** Build a **grouped bar chart** comparing 2 sub-factors (e.g. `GDP_per_capita` and `Freedom`) across the 5 most populated regions. Use colour meaningfully and write an insight title.

Regions to include: `'Western Europe'`, `'Latin America'`, `'East Asia'`, `'Sub-Saharan Africa'`, `'South Asia'`


```python
# Stretch goal — grouped bar chart
# YOUR CODE HERE
import pandas as pd
import plotly.express as px

# Regions specified in the exercise
regions = [
    'Western Europe',
    'Latin America and Caribbean',
    'East Asia',
    'Sub-Saharan Africa',
    'South Asia'
]

# Average GDP and Freedom by region
region_factors = (
    df[df['Region'].isin(regions)]
    .groupby('Region')[['GDP', 'Freedom']]
    .mean()
    .reset_index()
)

# Convert to long format for grouped bars
plot_df = region_factors.melt(
    id_vars='Region',
    value_vars=['GDP', 'Freedom'],
    var_name='Factor',
    value_name='Score'
)

fig = px.bar(
    plot_df,
    x='Region',
    y='Score',
    color='Factor',
    barmode='group',
    text='Score',
    color_discrete_map={
        'GDP': '#1f77b4',      # blue
        'Freedom': '#ff7f0e'   # orange
    }
)

fig.update_traces(
    texttemplate='%{text:.2f}',
    textposition='outside'
)

fig.update_layout(
    title={
        'text': 'Western Europe combines the strongest economic conditions with high personal freedom',
        'x': 0.5
    },
    xaxis_title='Region',
    yaxis_title='Average Score',
    template='simple_white',
    legend_title='',
    height=600
)

# Zero baseline (required for bar charts)
fig.update_yaxes(range=[0, plot_df['Score'].max() * 1.15])

fig.show()
```
